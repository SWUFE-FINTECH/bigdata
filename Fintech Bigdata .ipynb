## Part 2 : 大数据和小数据
### 2.1 小数据时代的数据获取方法
    **随机抽样** ：最少的数据获得最多的信息
    1）1890年美国的人口普查，数据汇总花费一年。
    2）收集和处理信息都不容易时，抽样是解决问题的较好办法
    统计学家证明，抽样分析的精确性随着抽样随机性的增加而大幅度提升，但与样本大小的增加关系不大。
    除了十年一次的大人口普查，美国人口普查局每年都会用随机抽样的方法对经济和人口进行200多次的小规模调查。
### 2.2 抽样调查方法的缺陷
     抽样方法的缺陷：
    1. 不好保证绝对随机性。
    2. 要求数据的绝对精确性
    3. 即使保证随机性，调查结果也存在3%的偏差。
    4. 随机抽样不适合考察子类的情况。
    5. 随机抽样需要详细的安排和执行。只可能从抽样数据中去验证实现设计好的问题结果。不可能有发现。
    
    现实实现绝对随机性很困难，一旦抽样过程中存在任何偏见，分析结果就会相去甚远。
    2008年，奥巴马和麦凯恩对决的总统大选。以固定电话用户为基础的民调，缺乏了随机性。民调结果出现3个百分点的偏差。

### 2.3 为什么我们需要大数据
    - 数据集越大，我们对问题的认识越清晰。
    - 数据算法的革命性改进，80年代的算法只能解决toy problem, 同样的算法今天可以解决复杂问题甚至接近人类的水平。

## Part 6： 大数据处理的基础设施

### 6.1 云计算
- 云计算是一种商业计算模型，它将计算任务分布在大量计算机构成的资源池上，使用用户能够按需获取计算力、存储空间和信息服务。
- 这个资源池称为“云”。通常是一些大型服务器集群，包括计算服务器、存储服务器等。云计算将计算资源集中起来，并通过专门的软件管理。用户可以动态的申请资源。
- 云计算是实施大数据分析的基础设施

### 6.2 云计算的特点
- 超大规模。云有相当的规模。当前全世界市场份额最大的三家云平台是Amazon的AWS，Microsoft Azure和Google Cloud platform。其中Azure在38个国家和地区有超过100个数据中心
- 虚拟化。云计算支持用户在任意位置，请求资源。具体请求的云资源来自哪台服务器，程序运行在那块CPU，用户无需知道。
- 高可靠性。“云”使用了数据多副本容错、计算节点同构可互换等措施来保证服务的高可靠性。使用云计算，比使用本地计算机更可靠。
- 高可伸缩性。“云”的规模可动态伸缩，满足应用和用户规模增长的需求。
- 客户按需购买。

### 6.4 大数据的处理软件架构
Hadoop
大数据处理的一个思路就是用计算机集群，提高数据处理能力。云计算的本质就是计算机的集群。当我们处理大数据时，如果自己构建集群来处理也是一个选项。Hadoop提供了这个可能性。

[Hadoop](https://hadoop.apache.org)

Hadoop是Apache开源组织的一个分布式计算框架，可以在大量廉价的硬件设备上组成集群，运行应用程序。Hadoop为应用程序提供了一组稳定可靠的接口。很多研究机构在用Hadoop构建云计算平台。

**Hadoop的核心：HDFS和MapReduce

Google公司面临信息爆炸的挑战开发出了GFS、MapReduce和Big Table等技术。
HDFS（Hadoop Distributed File System）是GFS的Hadoop实现版本。它解决了大数据存储的问题
MapReduce是海量数据的并行计算框架。解决了大数据的计算问题


#### 6.4.1 MapReduce的工作原理：

MapReduce的工作原理：
简单的说，MapReduce就是任务的分解与结果的汇总。

MapReduce采用分而治之的思想，把对大规模数据集进行操作，分发给一个主节点管理下的各分节点共同完成。然后通过整合各分节点的中间结果，得到最终结果。
上述过程被抽象成两个函数Map和Reduce。Map负责计算分解的任务，Reduce负责把处理结果汇总。
该任务必须要求带处理的数据集可以被分割成小的数据集，且可以并行计算。

### 6.6 Spark
#### 6.6.1 Hadoop在机器学习任务上的缺点
迭代式计算框架中，上一次迭代的输出（reduce产生的文件），是下一次迭代的输入。因为job是从磁盘读入文件，并写结果到磁盘。所以每次迭代都要从磁盘读入上一次迭代的输出结果，然后将本次迭代的结果写入磁盘。因为每次都要读磁盘，现在公认MapReduce写迭代的程序，效率低。

Spark是MapReduce的替代方案，而且兼容HDFS，可融入Hadoop的生态系统

#### 6.6.2 Spark的特点
Apache Spark是用于大规模数据处理的统一分析引擎。是在大数据集上运行机器学习算法的高效集群计算平台。它是由加州大学伯克利分校AMP实验室 开发的通用内存并行计算框架，用来构建大型的、低延迟的数据分析应用程序。它扩展了广泛使用的MapReduce计算模型。高效的支撑更多计算模式，包括交互式查询和流处理。spark的一个主要特点是能够在内存中进行计算，及时依赖磁盘进行复杂的运算，Spark依然比MapReduce更加高效。

Spark的四大特点：
- 运行速度比Hadoop提高100倍。
- Spark支持Java、Python和Scala的API，还支持超过80种高级算法，使用户可以快速构建不同的应用。
- Spark提供了统一的解决方案。Spark可以用于批处理、交互式查询（Spark SQL）、实时流处理（Spark Streaming）、机器学习（Spark MLlib）和图计算（GraphX）。
- 和Hadoop兼容。以处理所有Hadoop支持的数据。



## Part 7: 大数据分析技术
part6 讲物理层是数据的存储和实际运算。
逻辑层是各种算法

**机器学习的核心技术包含三大类：分类（包括预测）、聚类、Ensemble （集成技术）

学习问题可以粗略的分为有监督的和无监督的学习。

- **有监督的学习，训练数据集包含输入数据和对应的目标数据**。

当给出的训练数据包含的目标数据是类别标签，这是一个分类任务。当训练数据中给出的目标数据是连续的实数值，这是回归或预测任务。举例如下：

银行职员需要分析数据，搞清楚哪些贷款申请者是“合格的”。某电器商场的经理需要分析数据，以猜测哪些客户会购买一种新型的电脑。医学研究者希望分析癌症数据，预测病人应该接受三种治疗方案的哪一种。

上面描述的例子中，数据挖掘任务都是分类（classification），都需要构造一个模型或分类器（classifier）来预测类别标签。如贷款任务中的“安全”或“风险”；销售任务中的“是”或“否”；医疗例子中的“方案A”，“方案B”或“方案C”。这些类别标签之间的序没有意义。

假定上面的商场经理希望预测一位给定的客户在商场的一次促销活动中将花费多少钱。这是**数值预测**的例子。其中所构造的模型预测一个连续值函数或有序值，与类别标签不同。这种模型是预测器（predictor）。**回归分析**是数字预测最常用的统计学方法，因此，“预测”和“回归”这两个术语是同义词。

分类过程包含两个步骤：
1. 第一步，从训练集学习一个映射函数y=f(X)，即分类器。
2. 第二步，当有新的数据到来，使用该映射函数判断新数据的类别。预测过程与分类过程相同，只不过映射函数中的y是连续的实数。


- **无监督的学习中，训练集中的数据没有目标数据**。
当无监督学习的目标是发现数据集中相似的数据实例，这称为聚类任务。话题模型（Topic Model），例如，PLSA，LDA等，根据文档集合去学习其中谈论的话题这也是一个无监督学习的任务。

对于聚类任务的应用我们可以看上个例子。商场管理人员进行客户细分，将所有客户的客户划分到五个组，每个组应用不同的促销策略。而应用聚类算法，将所有客户聚为五个簇就可以完成客户细分的任务。

- **Ensemble(集成)，是数据挖掘中的一个重要技术**。
它的特点是不是采用单一的某个算法或模型完成分类，而是通过构造多个模型的组合，来建立一个性能更优的分类器。Ensemble技术在数据挖掘竞赛中经常取得很好的成绩。

数据挖掘中有很多的算法完成上述的分类、预测和聚类的任务。2006年ICDM国际会议上评选了数据挖掘的10大算法。它们包括：C4.5系列, K-Means，支持向量机，Apriori，EM算法，PageRank，AdaBoost，KNN，Naïve Bayes，CART。

### 7.5 模型的评估方法
当用训练集训练了一个模型，通常我们感兴趣模型在未见过的数据上性能表现如何。这决定了模型部署到真实世界时性能如何。我们用一个测试集来评价模型。测试集和训练集是分开的，即在训练模型时，测试集对模型是不开见的。模型在之前未观察到的数据集上表现的性能称为模型的**泛化（genralization）能力**。

我们使用训练集来训练模型。如果用训练集来评价模型，模型的误差称为**训练误差（train error）**。模型训练的过程是在不断的减小训练误差。训练误差很小说模型对数据的拟合很好，但不能证明模型的泛化性能。度量泛化能力的是**测试误差（test error）**，即模型对测试集上数据进行分类或预测，结果的误差的均值。

**越小的测试误差证明模型泛化能力越强**。

当我们手中有数据集，准备训练一个模型时，然后评估模型时，最常用的是K折交叉确认（cross validation）方法
我们把数据集划分成k份，用其中一份做测试集，剩下的k-1份做训练集。训练集训练模型，然后用测试集做评估。模型训练k次，每次用不同的一份数据集做测试集，其余的数据集做训练集。把k次测试的结果的均值作为最终模型的性能。

摸象大数据助力唯品会销售额的提升

### 8.1 广告投放策略 
根据客户关系发展的不同阶段，摸象大数据将唯品会用户群分成三大类：新用户、访问过但未产生购买的用户以及老用户。针对不同的用户群制定对应的个性化精准效果广告投放策略。 

### 8.2 新用户 
对于全新的用户，分析锁定唯品会目标受众，采用优惠券策略，通过首单优惠、红包等创意类型抓住用户眼球，吸引用户点击广告进入唯品会页面。

### 8.3 访问过但未产生购买的用户 
对于这部分用户，摸象大数据通过多次广告曝光，积累足够的分析数据进行分析筛选，目的是排除爬虫等虚假流量，大大降低唯品会的广告成本。 

### 8.4 老用户 
针对唯品会老用户，摸象大数据在人群标签体系基础上开展了更加细致的人群维度划分，区分用户长期品牌兴趣、短期购买欲望等，令广告有的放矢、千人千面，有效提高点击率和转化率。 




## Part 9 金融大数据的应用场景

大数据技术的应用提升了金融行业的资源配置效率，强化了风险管控能力。有效促进了金融行业的创新发展。金融大数据在银行业、证券行业、保险行业、支付清算行业和互联网金融都得到广泛的应用。

### Part 9.1 大数据在银行业的应用
信贷风险评估。
传统的方法中，银行对企业客户的违约风险评估是基于过往的信贷数据和交易数据等静态数据，这种方式的最大弊端就是缺乏前瞻性。因为影响企业的违约的重要因素并不仅仅是企业历史的信用情况，还包括行业的整体发展和实时的经营状况。而大数据手段的介入使信贷风险评估更趋近于事实


### Part 9.2 大数据在证券行业的应用

- 股市行情预测
大数据可以有效拓宽证券企业量化投资数据维度，帮助企业更精准的了解市场行情。随着大数据的广泛应用，数据规模爆发增长，以及数据分析、处理能力的显著提升，量化投资的模型性能将会得到显著的提升。

- 股价预测
诺贝尔经济学奖得主罗伯特.席勒设计的投资模型至今被业内沿用。在他的模型中，主要参考三个变量：投资项目计划的现金流、公司资本的估算成本、股票市场对投资的反映（市场情绪）。他认为，市场本身带有主观判断因素，投资者情绪会影响投资行为，而投资行为直接影响资产价格。然而，在大数据技术诞生前，市场情绪始终无法量化。大数据技术可以收集和分析社交网络的数据，了解市场对特点企业的观感，使得探知市场情绪成为可能。

- 智能投顾
智能投顾是最近几年证券公司应用大数据技术匹配客户多样化需求的新尝试之一，目前已成为财富管理新蓝海。智能投顾业务提供线上的投资顾问服务，能够基于客户的风险偏好、交易行为等个性化数据，采用量化模型为客户提供低门槛、低费率的个性化财富管理方案。
智能投顾在客户资料收集分析、投资方案的制定、执行以及后续的维护等步骤上均采用智能系统自动化的完成。

### Part 9.3 大数据在保险行业的应用

- 骗保识别
保险欺诈严重损害了保险公司的利益，为了识别可疑的保险欺诈案件，需要展开大量专项调查，不但耗费人力和财力，往往要持续数月甚至数年的时间。借助大数据手段，收集历史数据和各方位的数据集，建立保险欺诈模型，可以作为保险欺诈识别的有效辅助手段，提高保险欺诈识别的效率。

- 风险定价
保险企业对保费的定义是基于对一个群体的风险判断，对高风险的群体收取较高的费用，对低风险群体则降低费用。通过灵活的定价模式可以有效的提高客户的粘性。而大数据为这样的风险判断提供了前所未有的创新。
主要表现在，通过新的技术手段方便收集数据。例如新仪器设备采集行车数据，对车险的定价提供帮助。社交媒体上的信息对定义客户的性格，进而评估风险有帮助。

### Part 9.4 大数据在支付清算行业中的应用
- 交易欺诈识别
目前支付服务操作非常便捷，客户已经可以做到随时随地进行转账操作。面对盗刷和金融诈骗案件频发的现状，支付清算企业欺诈识别面临巨大挑战。
大数据可以利用账号基本信息、交易历史、位置信息、历史行为模式、正在发生行为模式，应用机器学习实时进行交易反欺诈分析。

### Part 9.5 大数据在互联网金融行业的应用
- 精准营销
- 黑产防范
互联网金融企业追求服务体验，强调便捷高效，简化手续。而这一特点容易被不法分子利用，虚假注册、利用网络购买的身份信息与银行卡套现，“多头借贷”乃至开发电脑程序骗贷已经形成一条黑色产业链，对互联网金融行业而言，欺诈风险高于信用风险。
大数据技术能够帮助企业掌握互联网金融黑产的特点，通过预警减少损失。
- 消费信贷
消费信贷和传统的企业信贷截然不同。它拥有小额、分散、高频、无抵押和利息跨度极大的特点。在贷款额度上可以小到100元；一家机构的一天放贷量可以达到上万到数十万笔。90%是纯信用贷，只能依靠数据进行审批。
大数据需要贯穿到客户生命周期。基于大数据的自动评分模型、自动审批系统和催收系统降低消费信贷的风险。






## Part 10: 大数据与法律风险
- 互联网公司就是技术公司，是数据公司。数据就是资产，是21世纪的石油，因此有人铤而走险。
- 有50%的利润，资本就会铤而走险；有100%的利润，资本就敢践踏人间一切法律—马克思
- 中国正在成为全球数据的最大源头。据IDC《数据时代2025》报告预测，中国数据总量占全球比重将从2018年的23.4%上升到2025年的27.8%，排名上升至第一位，绝对值将从7.6ZB上升到48.6ZB。但到目前为止，大部分人没有从自己源源不断生产的数据中获得任何报偿，反而招来了营销电话的骚扰，承担着隐私泄露的风险。
- 2017年6月1日，《网络安全法》正式实施，2018年被称为“中国数据合规元年”。

### 10.1 风险案例
今年3.15晚会上，央视用20分钟重点曝光了“骚扰电话产业链”，顺带曝光了背后非法获取、泄露、交易电话号、身份证等公民个人信息的黑产。
 
而就在晚会前夜，还发生了一件之后引发行业热议的事——号称拥有中国最全简历数据库的大数据公司巧达科技被北京警方“一锅端”。
其创始人说：
“我们的商业模式概括起来也就8个字——获取简历、数据变现。”

### 10.2 面临的数据风险

- 不当获取数据。2015年创立于深圳的车当，是一个提供二手车里程、维修等信息，抹平买卖双方信息不对称的车史信息平台。车当向中间人购买数据，滑向了危险。
- 数据安全问题。广义来说，安全既包括工作人员、合作方的不当操作或有意作恶导致的数据不当使用；也包括黑客攻击导致的数据损失和泄露。
- 数据授权问题。很多时候，用户选择同意授权是无奈之举，不同意就没法用服务，因而许多App的授权请求也被认为是霸王条款。
数据采集上是合法的，但整个生态却有不合理之处。这种风险不是单个公司的问题，而是全行业甚至全社会的共同难点。
